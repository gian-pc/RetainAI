## 1. Configuraci√≥n del Entorno

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import warnings

# Configuraci√≥n
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

# Seed para reproducibilidad
np.random.seed(42)

print("‚úÖ Librer√≠as importadas correctamente")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## 2. Carga de Datos

```python
# Rutas
DATA_PATH = Path('../../data/data.csv')
OUTPUT_PATH = Path('../../outputs/gian')

# Verificar que el archivo existe
if not DATA_PATH.exists():
    raise FileNotFoundError(f"Dataset no encontrado en: {DATA_PATH}")

print(f"üìÅ Cargando dataset desde: {DATA_PATH}")
print()
```

```python
# Cargar datos
df = pd.read_csv(DATA_PATH)

print("‚úÖ Dataset cargado exitosamente")
print(f"üìä Dimensiones: {df.shape[0]:,} registros √ó {df.shape[1]} columnas")
print(f"üíæ Memoria: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
```

---
## 3. Inspecci√≥n Inicial

```python
# Primeras filas
print("üìã Primeras 5 filas del dataset:")
print()
df.head()
```

```python
# Informaci√≥n general
print("‚ÑπÔ∏è  Informaci√≥n del dataset:")
print()
df.info()
```

```python
# Tipos de datos
print("üìä Distribuci√≥n de tipos de datos:")
print()
print(df.dtypes.value_counts())
print()
print(f"Num√©ricas: {len(df.select_dtypes(include=[np.number]).columns)}")
print(f"Categ√≥ricas: {len(df.select_dtypes(include=['object']).columns)}")
print(f"Fechas: {len(df.select_dtypes(include=['datetime']).columns)}")
```

```python
# Valores nulos
print("üîç An√°lisis de valores nulos:")
print()
null_counts = df.isnull().sum()
null_pct = (null_counts / len(df)) * 100

null_summary = pd.DataFrame({
    'Nulos': null_counts,
    'Porcentaje': null_pct
}).sort_values('Nulos', ascending=False)

# Mostrar solo columnas con nulos
null_summary[null_summary['Nulos'] > 0]
```

```python
# Duplicados
duplicates = df.duplicated().sum()
print(f"üîÑ Registros duplicados: {duplicates} ({duplicates/len(df)*100:.2f}%)")
```

## 4. Resumen Estad√≠stico

```python
# Estad√≠sticas descriptivas - Variables num√©ricas
print("üìà Estad√≠sticas descriptivas (variables num√©ricas):")
print()
df.describe()
```

```python
# Estad√≠sticas descriptivas - Variables categ√≥ricas
print("üìä Estad√≠sticas descriptivas (variables categ√≥ricas):")
print()
df.describe(include=['object'])
```

```python
# Variable target
if 'cancelacion' in df.columns:
    print("üéØ Distribuci√≥n de la variable target (cancelacion):")
    print()
    print(df['cancelacion'].value_counts())
    print()
    print("Porcentaje:")
    print(df['cancelacion'].value_counts(normalize=True) * 100)
    
    # Gr√°fico
    plt.figure(figsize=(8, 5))
    df['cancelacion'].value_counts().plot(kind='bar', color=['green', 'red'])
    plt.title('Distribuci√≥n de Churn', fontsize=14, fontweight='bold')
    plt.xlabel('Cancelaci√≥n (0=No, 1=S√≠)')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=0)
    plt.tight_layout()
    
    # Guardar gr√°fico
    output_fig = OUTPUT_PATH / 'figures' / 'eda' / '01_target_distribution.png'
    plt.savefig(output_fig, dpi=300, bbox_inches='tight')
    print(f"\n‚úÖ Gr√°fico guardado en: {output_fig}")
    
    plt.show()
```
