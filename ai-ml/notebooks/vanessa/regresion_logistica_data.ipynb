Definir target y eliminar columnas innecesarias

```python
TARGET = "cancelacion"

cols_to_drop = [
    "cliente_id",
    "fecha_registro",
    "fecha_ultimo_pago",
    "ultimo_contacto_soporte"
]

df = df.drop(columns=cols_to_drop, errors="ignore")

print(df.shape)
```

Separar X y y

```python
X = df.drop(columns=[TARGET])
y = df[TARGET]

print(X.shape, y.shape)
```

X train y X test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
```

```python
cat_cols = X_train.select_dtypes(include="object").columns.tolist()
num_cols = X_train.select_dtypes(exclude="object").columns.tolist()

print("Categóricas:", len(cat_cols))
print("Numéricas:", len(num_cols))
```

Preprocesamiento por imputación

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

num_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", num_pipeline, num_cols),
        ("cat", cat_pipeline, cat_cols)
    ]
)
```

#Modelo base

```python
from sklearn.linear_model import LogisticRegression

model = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("model", LogisticRegression(max_iter=1000, class_weight="balanced"))
])

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix

y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix, roc_auc_score

acc = accuracy_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_proba)
cm = confusion_matrix(y_test, y_pred)

print("Accuracy:", round(acc, 4))
print("Recall:", round(rec, 4))
print("ROC AUC:", round(auc, 4))
print("Matriz de confusión:\n", cm)
```

#Ajuste del modelo

```python
from sklearn.metrics import precision_score

thresholds = np.arange(0.05, 0.95, 0.05)

rows = []

for t in thresholds:
    y_pred_t = (y_proba >= t).astype(int)
    prec = precision_score(y_test, y_pred_t)
    rec = recall_score(y_test, y_pred_t)
    rows.append([t, prec, rec])

df_thr = pd.DataFrame(rows, columns=["threshold", "precision", "recall"])
df_thr
```

```python
t = 0.30
y_pred_30 = (y_proba >= t).astype(int)

from sklearn.metrics import confusion_matrix, precision_score, recall_score

cm_30 = confusion_matrix(y_test, y_pred_30)
prec_30 = precision_score(y_test, y_pred_30)
rec_30 = recall_score(y_test, y_pred_30)

print("Threshold:", t)
print("Precision:", round(prec_30, 4))
print("Recall:", round(rec_30, 4))
print("Matriz de confusión:\n", cm_30)
```

Obtener probabilidades de churn

```python
y_prob = model.predict_proba(X_test)[:, 1]
```

Curva Precision–Recall

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
```

Tabla de trade-off

```python
pr_df = pd.DataFrame({
    "threshold": thresholds,
    "precision": precision[:-1],
    "recall": recall[:-1]
})

pr_df.head()
```

```python
best_threshold = 0.17

y_pred_custom = (y_prob >= best_threshold).astype(int)

print("Accuracy:", accuracy_score(y_test, y_pred_custom))
print("Recall:", recall_score(y_test, y_pred_custom))
confusion_matrix(y_test, y_pred_custom)
```

Reajuste de modelo

```python
t = 0.40
y_pred_40 = (y_proba >= t).astype(int)

cm_40 = confusion_matrix(y_test, y_pred_40)
prec_40 = precision_score(y_test, y_pred_40)
rec_40 = recall_score(y_test, y_pred_40)

print("Threshold:", t)
print("Precision:", round(prec_40, 4))
print("Recall:", round(rec_40, 4))
print("Matriz de confusión:\n", cm_40)

print("Clientes en riesgo:", cm_40[0,1] + cm_40[1,1])
print("Porcentaje base:", (cm_40[0,1] + cm_40[1,1]) / len(y_test))
```
